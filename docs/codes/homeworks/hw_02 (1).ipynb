# Análisis del Impacto COVID-19 en el Aprendizaje Digital

Este notebook explora cómo la pandemia afectó el uso de plataformas educativas digitales en Estados Unidos durante 2020, analizando la relación entre participación estudiantil, características demográficas de los distritos y acceso a internet.

## 1. Importar librerías necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import warnings
warnings.filterwarnings('ignore')

# Configuración de gráficos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline
```

## 2. Cargar los datos

Vamos a cargar los tres componentes principales del dataset:
- Información de distritos escolares
- Información de productos educativos
- Datos de engagement (participación)

```python
# Cargar información de distritos
districts = pd.read_csv('input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
print(f"Distritos cargados: {len(districts)} registros")
print(f"Columnas: {list(districts.columns)}")
print(districts["pct_free/reduced"])

# Cargar información de productos
products = pd.read_csv('input/learnplatform-covid19-impact-on-digital-learning/products_info.csv')
print(f"\nProductos cargados: {len(products)} registros")

# Cargar datos de engagement
engagement_files = glob.glob('input/learnplatform-covid19-impact-on-digital-learning/engagement_data/*.csv')
print(f"\nArchivos de engagement encontrados: {len(engagement_files)}")

# Combinar todos los archivos de engagement con su district_id
engagement_list = []
for file in engagement_files:
    df_temp = pd.read_csv(file)
    # Extraer el district_id del nombre del archivo
    district_id = file.split("\\")[-1].split(".")[0]  # usa "\\" porque estás en Windows
    df_temp["district_id"] = district_id
    engagement_list.append(df_temp)

engagement = pd.concat(engagement_list, ignore_index=True)
print(f"Total registros de engagement: {len(engagement)}")
print(f"Columnas: {list(engagement.columns)}")
```

## 3. Exploración inicial de los datos

```python
# Ver estructura de los datos de distritos
print("=== INFORMACIÓN DE DISTRITOS ===")
print(len(districts))
print(districts.head())
print(f"\nValores nulos:\n{districts.isnull().sum()}")
```

El pp_total_raw se corresponde a el gasto local + gasto federal en la escuela, es bastante normal que haya tantos datos faltates.

El pct_black/hispanic es el porcentaje de estudiantes hispanicos o negros, una vez más, como no es una medición que se haga siempre, es normal que falten muchos datos.

```python
# Ver estructura de engagement
print("=== DATOS DE ENGAGEMENT ===")
print(engagement.head())
print(f"\nColumnas: {list(engagement.columns)}")
print(f"\nValores nulos:\n{engagement.isnull().sum()}")
```

## 4. Limpieza y preparación de datos

Vamos a limpiar los datos y crear un DataFrame maestro que combine toda la información relevante.

```python
# Convertir la columna de tiempo a datetime
engagement['time'] = pd.to_datetime(engagement['time'])

# Extraer mes para análisis temporal
engagement['month'] = engagement['time'].dt.month
engagement['year'] = engagement['time'].dt.year

# Filtrar solo datos de 2020 (año de la pandemia)
engagement_2020 = engagement[engagement['year'] == 2020].copy()

print(f"Registros en 2020: {len(engagement_2020)}")
print(f"Rango de fechas: {engagement_2020['time'].min()} a {engagement_2020['time'].max()}")
```

```python
print("Engagement columns:", engagement_2020.columns.tolist())
print("Districts columns:", districts.columns.tolist())
```

```python
# Asegurar que ambas columnas 'district_id' sean del mismo tipo (string)
engagement['district_id'] = engagement['district_id'].astype(str)
districts['district_id'] = districts['district_id'].astype(str)

# Crear DataFrame maestro
master_df = engagement.merge(districts, on='district_id', how='left')
master_df = master_df.merge(products, left_on='lp_id', right_on='LP ID', how='left')

print(f"DataFrame maestro creado: {len(master_df)} registros")
print(f"Columnas finales: {len(master_df.columns)}")

# Ver primeras filas del dataset combinado
master_df.head()
```

### 5.1 Estadísticas descriptivas básicas

```python
# Estadísticas de engagement
print("=== ESTADÍSTICAS DE ENGAGEMENT ===")
print(master_df['engagement_index'].describe())

# Conteo por tipo de localidad
if 'locale' in master_df.columns:
    print("\n=== DISTRIBUCIÓN POR TIPO DE LOCALIDAD ===")
    print(master_df['locale'].value_counts())
```

### 5.2 Análisis temporal del engagement

Veamos cómo evolucionó la participación a lo largo de 2020.

```python
# Calcular engagement promedio por mes
monthly_engagement = master_df.groupby('month')['engagement_index'].agg(['mean', 'count']).reset_index()

# Gráfico de tendencia mensual
fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.plot(monthly_engagement['month'], monthly_engagement['mean'], marker='o', linewidth=2, markersize=8, color='#2E86AB')
ax1.set_xlabel('Mes (2020)', fontsize=11)
ax1.set_ylabel('Engagement Promedio', fontsize=11, color='#2E86AB')
ax1.tick_params(axis='y', labelcolor='#2E86AB')
ax1.grid(True, alpha=0.3)
ax1.set_xticks(range(1, 13))
ax1.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

# Marcar inicio de pandemia (marzo 2020)
ax1.axvline(x=3, color='red', linestyle='--', alpha=0.5, linewidth=2)
ax1.text(3.1, ax1.get_ylim()[1]*0.95, 'Inicio pandemia\n(marzo 2020)', fontsize=9, color='red')

plt.title('Evolución del Engagement en Plataformas Digitales durante 2020', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("Observaciones:")
print(f"- Engagement más alto: Mes {monthly_engagement.loc[monthly_engagement['mean'].idxmax(), 'month']:.0f}")
print(f"- Engagement más bajo: Mes {monthly_engagement.loc[monthly_engagement['mean'].idxmin(), 'month']:.0f}")
```

Parece ser que la participación aumentó cuando dio inicio la pandemia, pero se reduce drásticamente en julio, esto es casi seguro que se debe a las vacaciones de verano en estados unidos. Luego, cuando se vuelve de vacaciones de verano, se vuelve más fuerte que nunca, para tener una disminución gradual de el engagement observado a medida que pasan los meses.

### 5.4 Relación entre acceso a banda ancha y engagement

Analicemos si los distritos con mejor acceso a internet tienen mayor participación.

```python
# Buscar columnas relacionadas con banda ancha/internet/demografía
broadband_cols = [col for col in master_df.columns if 'broadband' in col.lower() or 'internet' in col.lower() or 'pp_' in col.lower()]
print(f"Columnas de conectividad encontradas: {broadband_cols}")

# Buscar columnas numéricas que podrían ser relevantes
numeric_demo_cols = []
for col in master_df.columns:
    if master_df[col].dtype in ['float64', 'int64'] and col not in ['district_id', 'lp_id', 'LP ID', 'year', 'month', 'engagement_index']:
        # Verificar que tenga variabilidad
        if master_df[col].nunique() > 10:
            numeric_demo_cols.append(col)

print(f"\nColumnas numéricas disponibles para análisis: {numeric_demo_cols[:10]}")  # Mostrar solo las primeras 10

# Intentar análisis con la primera columna numérica válida
if len(numeric_demo_cols) > 0:
    # Usar la primera columna numérica encontrada
    demo_col = numeric_demo_cols[0]
    print(f"\nUsando columna: {demo_col} para análisis demográfico")

    # Eliminar valores nulos
    df_temp = master_df[[demo_col, 'engagement_index']].dropna()

    if len(df_temp) > 20:
        try:
            # Crear bins para categorizar
            df_temp['demo_category'] = pd.qcut(df_temp[demo_col], q=4, labels=['Bajo', 'Medio-Bajo', 'Medio-Alto', 'Alto'], duplicates='drop')

            # Engagement por categoría
            demo_engagement = df_temp.groupby('demo_category')['engagement_index'].mean()

            plt.figure(figsize=(10, 5))
            demo_engagement.plot(kind='bar', color='#06A77D', edgecolor='black', linewidth=1.2)
            plt.xlabel(f'Nivel de {demo_col}', fontsize=11)
            plt.ylabel('Engagement Promedio', fontsize=11)
            plt.title(f'Relación entre {demo_col} y Engagement Digital', fontsize=13, fontweight='bold')
            plt.xticks(rotation=0)
            plt.grid(axis='y', alpha=0.3)
            plt.tight_layout()
            plt.show()

            # Calcular correlación
            corr = df_temp[[demo_col, 'engagement_index']].corr().iloc[0, 1]
            print(f"\nCorrelación entre {demo_col} y engagement: {corr:.3f}")

            if abs(corr) > 0.3:
                print(f"✓ Se observa una correlación {'positiva' if corr > 0 else 'negativa'} fuerte")
            elif abs(corr) > 0.1:
                print(f"→ Se observa una correlación {'positiva' if corr > 0 else 'negativa'} débil")
            else:
                print("→ No se observa correlación significativa")

        except Exception as e:
            print(f"No se pudo crear el análisis: {e}")
            print("Continuando con el resto del análisis...")
else:
    print("\nNo se encontraron columnas demográficas numéricas válidas.")
    print("Saltando este análisis y continuando...")
```

Exactamente como uno esperaría, los estudiantes que no abrían ninguna pestaña de un producto dado en el día son los que tienen una participación casi nula. Esto puede hacer que sea una variable poco interensate y que sea más o menos equivalente a que el engagement sea igual o distinto de 0.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Obtener datos sin nulos
engagement_data = master_df['engagement_index'].dropna()

# Histograma con escala logarítmica en Y para ver mejor la distribución
axes[0].hist(engagement_data, bins=50, color='#577590', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Engagement Index', fontsize=11)
axes[0].set_ylabel('Frecuencia (escala log)', fontsize=11)
axes[0].set_title('Distribución del Engagement', fontsize=12, fontweight='bold')
axes[0].set_yscale('log')
axes[0].grid(axis='y', alpha=0.3)

# Box plot también con escala logarítmica
bp = axes[1].boxplot(engagement_data, vert=True, patch_artist=True,
                boxprops=dict(facecolor='#43AA8B', alpha=0.7),
                medianprops=dict(color='red', linewidth=2))
axes[1].set_ylabel('Engagement Index (escala log)', fontsize=11)
axes[1].set_title('Box Plot del Engagement', fontsize=12, fontweight='bold')
axes[1].set_yscale('log')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Detectar outliers
Q1 = engagement_data.quantile(0.25)
Q3 = engagement_data.quantile(0.75)
IQR = Q3 - Q1
outliers = master_df[(master_df['engagement_index'] < Q1 - 1.5*IQR) | (master_df['engagement_index'] > Q3 + 1.5*IQR)]

print(f"\n=== ANÁLISIS DE LA DISTRIBUCIÓN ===")
print(f"Media: {engagement_data.mean():.3f}")
print(f"Mediana: {engagement_data.median():.3f}")
print(f"Desviación estándar: {engagement_data.std():.3f}")
print(f"Mínimo: {engagement_data.min():.3f}")
print(f"Máximo: {engagement_data.max():.3f}")

print(f"\nOutliers detectados: {len(outliers)} ({len(outliers)/len(master_df)*100:.1f}% del total)")

# Contar cuántos valores están cerca de cero
near_zero = (engagement_data < 0.1).sum()
print(f"Valores cercanos a cero (< 0.1): {near_zero} ({near_zero/len(engagement_data)*100:.1f}%)")

print("\n📊 INTERPRETACIÓN:")
print("La distribución está altamente sesgada hacia valores bajos, lo cual es típico en datos")
print("de engagement educativo donde:")
print("  • Muchas plataformas tienen uso ocasional o nulo")
print("  • Pocos productos concentran la mayor parte del uso")
print("  • Existe una 'cola larga' de productos con alto engagement")
```

## 6. Análisis de Brechas Digitales

Identifiquemos las principales brechas en acceso y uso de tecnología educativa.

```python
# Comparar engagement antes y después del inicio de la pandemia (marzo 2020)
pre_pandemic = master_df[master_df['month'] < 3]['engagement_index'].mean()
post_pandemic = master_df[master_df['month'] >= 3]['engagement_index'].mean()

print("=== IMPACTO DE LA PANDEMIA ===")
print(f"Engagement promedio pre-pandemia (Ene-Feb): {pre_pandemic:.3f}")
print(f"Engagement promedio post-pandemia (Mar-Dic): {post_pandemic:.3f}")
print(f"Cambio: {((post_pandemic/pre_pandemic - 1) * 100):+.1f}%")

# Visualizar el impacto
plt.figure(figsize=(8, 5))
categories = ['Pre-Pandemia\n(Ene-Feb)', 'Post-Pandemia\n(Mar-Dic)']
values = [pre_pandemic, post_pandemic]
colors = ['#4ECDC4', '#FF6B6B']
bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5, width=0.6)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.ylabel('Engagement Promedio', fontsize=11)
plt.title('Comparación de Engagement: Antes vs Durante la Pandemia', fontsize=13, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

Notamos que el uso de productos digitales subió casi en un 50% después de la pandemia.

## 6.5 Comparación pct_free/reduced con engagement ##

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

## Cargar datos

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(f"Train: {train.shape}, Test: {test.shape}")
print(f"\nTasa de supervivencia: {train['Survived'].mean():.2%}")
train.head()
```

## Exploración rápida

```python
print("Valores faltantes:")
print(train.isnull().sum())
print("\nSupervivencia por género:")
print(train.groupby('Sex')['Survived'].mean())
print("\nSupervivencia por clase:")
print(train.groupby('Pclass')['Survived'].mean())
```

## Preparación de datos

```python
def preparar_datos(df):
    df = df.copy()

    # Rellenar Age con la mediana
    df['Age'].fillna(df['Age'].median(), inplace=True)

    # Rellenar Embarked con la moda
    df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

    # Rellenar Fare con la mediana
    df['Fare'].fillna(df['Fare'].median(), inplace=True)

    # Crear nuevas características
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

    # Convertir Sex a numérico
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

    # Convertir Embarked a numérico
    df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

    return df

train_prep = preparar_datos(train)
test_prep = preparar_datos(test)
```

## Seleccionar características

```python
features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize', 'IsAlone']

X = train_prep[features]
y = train_prep['Survived']
X_test = test_prep[features]

print(f"Características seleccionadas: {features}")
```

## Entrenar modelo

```python
modelo = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)

# Validación cruzada
scores = cross_val_score(modelo, X, y, cv=5)
print(f"Accuracy promedio (CV): {scores.mean():.4f} (+/- {scores.std():.4f})")

# Entrenar con todos los datos
modelo.fit(X, y)
print("\nModelo entrenado exitosamente!")
```

## Generar predicciones

```python
predicciones = modelo.predict(X_test)

# Crear archivo de submission
submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predicciones
})

submission.to_csv('submission.csv', index=False)

print("Archivo 'submission.csv' generado exitosamente!")
print(f"\nPredicciones de supervivencia: {predicciones.sum()} de {len(predicciones)}")
submission.head(10)
```

```python
# --- MODULES ---
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


# --- DATA WRANGLING ---
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
passengerId = test.PassengerId

titanic = pd.concat([train, test], ignore_index=True)
train_idx = len(train)
test_idx = len(titanic) - len(test)

# --- FEATURE ENGINEERING ---
titanic['Title'] = titanic.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())

standardized_titles = {
    "Capt": "Officer", "Col": "Officer", "Major": "Officer",
    "Jonkheer": "Royalty", "Don": "Royalty", "Sir": "Royalty",
    "Dr": "Officer", "Rev": "Officer", "the Countess": "Royalty",
    "Dona": "Royalty", "Mme": "Mrs", "Mlle": "Miss", "Ms": "Mrs",
    "Mr": "Mr", "Mrs": "Mrs", "Miss": "Miss", "Master": "Master", "Lady": "Royalty"
}
titanic['Title'] = titanic['Title'].map(standardized_titles)

grouped = titanic.groupby(['Sex', 'Pclass', 'Title'])
titanic['Age'] = grouped['Age'].transform(lambda x: x.fillna(x.median()))


titanic['Cabin'] = titanic['Cabin'].fillna('U')
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].value_counts().index[0])
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())

titanic['FamilySize'] = titanic['Parch'] + titanic['SibSp'] + 1
titanic['Cabin'] = titanic['Cabin'].map(lambda x: x[0])

# --- DUMMIES AND CLEANUP ---
titanic['Sex'] = titanic['Sex'].map({"male": 0, "female": 1})
pclass_dummies = pd.get_dummies(titanic['Pclass'], prefix="Pclass")
title_dummies = pd.get_dummies(titanic['Title'], prefix="Title")
cabin_dummies = pd.get_dummies(titanic['Cabin'], prefix="Cabin")
embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix="Embarked")

titanic_dummies = pd.concat(
    [titanic, pclass_dummies, title_dummies, cabin_dummies, embarked_dummies], axis=1
)
titanic_dummies.drop(['Pclass', 'Title', 'Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)

# --- TRAIN / TEST SPLIT ---
train = titanic_dummies[:train_idx]
test = titanic_dummies[test_idx:]

train['Survived'] = train['Survived'].astype(int)
X = train.drop('Survived', axis=1).values
y = train['Survived'].values
X_test = test.drop('Survived', axis=1).values

# --- MODELING ---
modelo = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
modelo.fit(X, y)


# --- PREDICTION ---
forrest_pred = modelo.predict(X_test)

# --- Guardar submission ---
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': forrest_pred})
submission.to_csv('submission3.csv', index=False)

```
